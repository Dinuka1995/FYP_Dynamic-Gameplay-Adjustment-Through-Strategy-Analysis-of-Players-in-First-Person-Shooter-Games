```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
import random

%matplotlib notebook
```

```python
df = pd.read_csv("./data_set_new.csv")

columns = list(df.columns.values)

fig = plt.figure(figsize=(10,20), subplotpars=matplotlib.figure.SubplotParams(hspace=0.25))
for i in range(len(columns)-1):
    x1 = list(df[df["is_aggressive"]==0][columns[i]])
    y1 = list(df[df["is_aggressive"]==0]["is_aggressive"])

    x2 = list(df[df["is_aggressive"]==1][columns[i]])
    y2 = list(df[df["is_aggressive"]==1]["is_aggressive"])

    data = ((x1,y1), (x2,y2))
    colors = ("red", "blue")
    groups = ("passive", "aggressive")

    # Create plot
    ax = fig.add_subplot(5, 2, i+1)

    for data, color, group in zip(data, colors, groups):
        x, y = data
        ax.scatter(x, y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

    plt.title('\''+ columns[i] + '\' vs \'is aggresive\'')
    plt.legend(loc=5)
plt.show(block=True)
```

```python
fig = plt.figure(figsize=(10,80), subplotpars=matplotlib.figure.SubplotParams(hspace=0.25))
k = 1
for i in range(len(columns)-1):
    for j in range(i+1, len(columns)-1):
        x1 = list(df[df["is_aggressive"]==0][columns[i]])
        y1 = list(df[df["is_aggressive"]==0][columns[j]])

        x2 = list(df[df["is_aggressive"]==1][columns[i]])
        y2 = list(df[df["is_aggressive"]==1][columns[j]])

        data = ((x1,y1), (x2,y2))
        colors = ("red", "blue")
        groups = ("passive", "aggressive")

        # Create plot
        ax = fig.add_subplot(18, 2, k)

        for data, color, group in zip(data, colors, groups):
            x, y = data
            ax.scatter(x, y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

        plt.title('\''+ columns[i] + '\' vs \'' + columns[j] + '\'')
        plt.xlabel(columns[i])
        plt.ylabel(columns[j])
        k+=1
plt.legend(loc=5)
plt.show(block=True)
```

```python
im_features = [[8,5,6], [8,5,7], [8,6,7], [5,6,7]]
fig = plt.figure(figsize=(10,40))
l=1
for f_list in im_features:
    i = f_list[0]
    j = f_list[1]
    k = f_list[2]
    x1 = list(df[df["is_aggressive"]==0][columns[i]])
    y1 = list(df[df["is_aggressive"]==0][columns[j]])
    z1 = list(df[df["is_aggressive"]==0][columns[k]])

    x2 = list(df[df["is_aggressive"]==1][columns[i]])
    y2 = list(df[df["is_aggressive"]==1][columns[j]])
    z2 = list(df[df["is_aggressive"]==1][columns[k]])

    data = ((x1,y1,z1), (x2,y2,z2))
    colors = ("red", "blue")
    groups = ("passive", "aggressive")
    
    ax = fig.add_subplot(4, 1, l, projection='3d')
    # ax = Axes3D(ax_)

    for data, color, group in zip(data, colors, groups):
        x, y, z = data
        ax.scatter(x, y, z, c=color, label=group)

    plt.title('\''+ columns[i] + '\' vs \'' + columns[j] + '\' vs \'' + columns[k] + '\'')
    ax.set_xlabel(columns[i])
    ax.set_ylabel(columns[j])
    ax.set_zlabel(columns[k])
    l+=1

plt.legend()
plt.show()
```
